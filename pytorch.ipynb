### **Tensor Math Operations<br>**
* Add , Subtract, Multiply , Divide , Remainders , Exponents <br>
* ShortHand and uh , LongHand<br>
* Reassignment<br>

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
## Addition
tensor_a + tensor_b
```

```python
# Addition longhand
torch.add(tensor_a,tensor_a)
```

```python
#Subtraction
tensor_b - tensor_a
```

```python
#  LongHand
torch.sub(tensor_b,tensor_a)
```

```python
# Multiplication
tensor_a * tensor_b
```

```python
torch.mul(tensor_a,tensor_b)
```

```python
# Division
tensor_b / tensor_a
```

```python
# LongHand
torch.div(tensor_b,tensor_a)
```
